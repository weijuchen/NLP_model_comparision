```python
data = []
pd.set_option("display.max_colwidth", None)


with open("test.ft.txt", "r", encoding="utf-8") as f:
    for line in f:
        result=line.strip()   # 去除\n
        label,text =result.split(" ",1) # 用空白分割一次
        label = label.replace("__label__", "")  # 去除不需要的字
        data.append([label,text])


print(data[0])
```

```python
# turn it into a dataframe
data_df=pd.DataFrame(data,columns=["rank","sentence"])
data_df.head(3)

# make a copy of the dataframe
df=data_df.copy()
df.head(3)
```

```python
# import the VADER sentiment library
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
text = df.sentence[0]
text
```

```python
# create an analyzer object, apply it to the text data and view the polarity scores
analyzer=SentimentIntensityAnalyzer()
res=analyzer.polarity_scores(text)
print(res)
print(res['compound'])

```

```python
analyzer=SentimentIntensityAnalyzer()
def sentiment_intensity_score(text):
    score=analyzer.polarity_scores(text)
    return score['compound']

df['compound_score']=df['sentence'].apply(sentiment_intensity_score)
# compound_score=df['sentence'].apply(sentiment_intensity_score)  wrong way
df
# df['compound_score'].loc[:2]
```

```python
df = df.rename(columns={'compound_score': 'Sentiment_VADER'})
df.head(5)
```

```python
from transformers import pipeline


# calculate the sentiment scores
sentiment_analyzer = pipeline("sentiment-analysis",
                              model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",
                              device=0,  # -1 means CPU
                              framework="pt",
                              truncation=True,  # adding truncation here to truncate text before analyzing sentiment
                               use_fast=True # 2. faster tokenization)
                           )
sentiment_scores = df['sentence'].apply(sentiment_analyzer)
sentiment_scores[:5]
```

```python
# extract the label for a single review
sentiment_scores[0][0]['label']
```

```python
# extract the score for a single review
sentiment_scores[0][0]['score']
```

```python
# extract the label and score and create a sentiment score for all reviews
df['Label_HF'] = sentiment_scores.apply(lambda x: x[0]['label'])
df['Score_HF'] = sentiment_scores.apply(lambda x: x[0]['score'])
df['Sentiment_HF'] = df.apply(lambda row: row['Score_HF'] if row['Label_HF'] == 'POSITIVE' else -row['Score_HF'], axis=1)
```
