```python
import pandas as pd
import csv
```

```python
data = []
pd.set_option("display.max_colwidth", None)


with open("test.ft.txt", "r", encoding="utf-8") as f:
    for line in f:
        result=line.strip()   # 去除\n
        label,text =result.split(" ",1) # 用空白分割一次
        label = label.replace("__label__", "")  # 去除不需要的字
        data.append([label,text])


print(data[0])
```

```python
# turn it into a dataframe
data_df=pd.DataFrame(data,columns=["rank","sentence"])
data_df.head(3)

# make a copy of the dataframe
df=data_df.copy()
df.head(3)
```

```python
# import the VADER sentiment library
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
text = df.sentence[0]
text
```

```python
# create an analyzer object, apply it to the text data and view the polarity scores
analyzer=SentimentIntensityAnalyzer()
res=analyzer.polarity_scores(text)
print(res)
print(res['compound'])

```

```python

analyzer=SentimentIntensityAnalyzer()
def sentiment_intensity_score(text):
    score=analyzer.polarity_scores(text)
    return score['compound']

df['compound_score']=df['sentence'].apply(sentiment_intensity_score) 
# compound_score=df['sentence'].apply(sentiment_intensity_score)  wrong way
df
# df['compound_score'].loc[:2]

    
```
